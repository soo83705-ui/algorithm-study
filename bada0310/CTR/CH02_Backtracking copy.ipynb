```python
# greedy  + use lambda 
N = int(input())
segments = []
for _ in range(N):
    s, e = map(int,input().split())
    segments.append((s,e))
    
segments.sort(key=lambda x: (x[1],x[0]))
cnt = 0 # count
curr_end = -1 # 마지막으로 고른 선분이 끝나는 지점을 저장할 val

for end, start in segments:
    if start > curr_end:
        cnt += 1
        curr_end = end
        
print(cnt)
```

```python
# greedy 
# 겹치지 않는 선분의 길이 
N = int(input())
segments = []
for _ in range(N):
    s, e = map(int,input().split())
    segments.append((e,s))

segments.sort()
cnt = 0 # count
curr_end = -1 # 마지막으로 고른 선분이 끝나는 지점을 저장할 val

for end, start in segments:
    if start > curr_end:
        cnt += 1
        curr_end = end
        
print(cnt)
```

```python
# backtracking 
def backtrack(idx, curr_end, cnt):
    global max_cnt
    
    if idx == N: # if i travel all nodes 
        max_cnt = cnt
        return
    
    start,end =segments[idx]
    # if i choose this node
    if start > curr_end:
        backtrack(idx+1,end, cnt+1)
    # if i not chooese this node
    backtrack(idx+1, curr_end, cnt)

N = int(input())
for _ in range(N):
    start, end = map(int,input().split())
    segments.append((start,end))
    segments.sort()
    
    max_cnt = 0
    backtrack(0,-1, 0)
    print(max_cnt)
```

```python
# 1931 회의실 배정 
# N(1 ≤ N ≤ 100,000)
def backtrack(idx, curr_end, cnt):
    global max_cnt
    if idx == N:
        max_cnt = max(max_cnt,cnt)
        return
    start, end = segments[idx]
    
    if start >= curr_end:
        backtrack(idx+1, end, cnt+1)
    backtrack(idx+1,curr_end,cnt)
        

N = int(input())
segments = []
for _ in range(N):
    start, end = map(int,input().split())
    segments.append((start,end))

segments.sort() # start straight
max_cnt = 0
backtrack(0,-1,0)
print(max_cnt)
```
